```scala
//Find the top movies with highest average rating

val ratings = sc.textFile("/user/itv004483/sparkinput/ratings.dat")

val mapRatings = ratings.map( x=> (x.split("::")(1), x.split("::")(2)))
//.filter(x=> x._1 == "2905")

//mapRatings.collect.foreach(println)


val mapRatingCnt = mapRatings.mapValues(x => (x.toFloat,1.0))

val mapUsrCnt = mapRatingCnt.reduceByKey( (x,y) => (x._1+y._1 , x._2+y._2))
//mapUsrCnt.collect.foreach(println)

val avgRatings = mapUsrCnt.filter( x=> x._2._2 > 10).mapValues( x=> (x._1/x._2))

val topRatings = avgRatings.filter( x => x._2 > 4)

val moviesData = sc.textFile("/user/itv004483/sparkinput/movies.dat")
val movies = moviesData.map(x => (x.split("::")(0), x.split("::")(1)))

val joinrdd = topRatings.join(movies)
//joinrdd.collect.foreach(println)

val res = joinrdd.map(x => (x._2._2,x._2._1)).sortBy(x=> x._2, ascending = false)
res.collect.foreach(println)
```
